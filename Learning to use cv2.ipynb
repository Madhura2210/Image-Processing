Convert rgb to grayscale

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("window",img_gray)
cv2.waitKey(0)
```

```python
print(img_gray.shape)
```

We see how we have to take arguments in reverse while resizing [1] and then [0]th index has to be passed

We also have to take into account rgb is bgr in cv2, it is all reversed

```python
new_img = img
new_img[:,:,0]=0
cv2.imshow("window",new_img)
cv2.waitKey(0) # we remove blue 0 is Blue
```

```python
new_img_1 = img
new_img_1[:,:,1]=0
cv2.imshow("window",new_img_1)
cv2.waitKey(0) # we remove green 1 is green
```

```python
new_img_2= img
new_img_2[:,:,2]=0
cv2.imshow("window",new_img_2)
cv2.waitKey(0) # we remove red BGR order 2 is red
```

```python
#drawing shapes
img_new = np.zeros((512,512,3))
cv2.imshow("window",img_new)
cv2.waitKey(0)
```

```python
img_1=cv2.rectangle(img_new,pt1=(100,100),pt2=(300,300),color=(0,130,0),thickness=-1)
cv2.imshow("window",img_1)
cv2.waitKey(0)
```

```python
img_1=cv2.circle(img_new,center=(100,100),radius=50,color=(0,0,100),thickness=-1)
cv2.imshow("window",img_1)
cv2.waitKey(0)
```

```python
img_1=cv2.line(img_new,pt1=(100,100),pt2=(300,100),color=(100,100,0))
cv2.imshow("window",img_1)
cv2.waitKey(0)
```

```python
img_1=cv2.putText(img_new,org=(100,200),fontScale=1,color=(0,255,130),thickness=2,lineType=cv2.LINE_AA,text="Hello",fontFace=cv2.FONT_ITALIC)
cv2.imshow("window",img_1)
cv2.waitKey(0)
```
